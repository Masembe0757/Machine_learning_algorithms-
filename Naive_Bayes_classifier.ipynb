```python
#CALLED NAIVE BECAUSE WE ASSUME INDEPENDENCE AMONG FEATURES OF THE DATASET
#Good for grouping and categorising e.g spam detection
import pandas as pd
```

```python
df = pd.read_csv('covid data.csv')
```

```python
df.STATUS = df.STATUS.apply(lambda x : x.replace('Died','0').replace('Cured','1'))
df
```

```python
#Removing unknown values
for x in df.columns:
    df = df[df[x]!=97]
    df = df[df[x]!=98]
    df= df[df[x] != 99]
df
```

```python
x = df[['INTUBED', 'PNEUMONIA', 'AGE', 'DIABETES', 'ASTHMA', 'CARDIOVASCULAR',
       'OBESITY', 'TOBACCO', 'ICU']]
y = df.STATUS
```

```python
from sklearn.model_selection import train_test_split as ttt
X_train,X_test,y_train,y_test = ttt(x,y ,test_size = 0.1)
```

```python
#There are many bayes model classes but gaussian is best for normal distribution datasets, Bernouli for binary and multinomial for descrete 
#data such as rating of a movie from5 to five
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#71% accurate
```

```python
df2 = pd.read_csv('spam.csv')
df2
```

```python
df2.describe()
```

```python
df2['spam'] = df2.Category.apply(lambda x : 1 if x == 'spam' else 0 ) 
```

```python
df2
```

```python
from sklearn.model_selection import train_test_split as ttt
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X_train,X_test,y_train,y_test = ttt(df2.Message,df2.spam, test_size = 0.2 , random_state = 42)
```

```python
#USing a pipeline
from sklearn.pipeline import Pipeline
pln = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())

])
```

```python
pln.fit(X_train,y_train)
```

```python
pln.score(X_test,y_test)
```

```python
received = (
    [
        'Helo Mr. Petersone was inquiring about pending tuition payments',
        'Helo you have won a brand new Iphone 15, click here to claim your prize'
    ])
```

```python
pln.predict(received)
```
