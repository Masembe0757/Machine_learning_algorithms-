```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('covid data.csv')
#Removing unknown values
for x in df.columns:
    df = df[df[x]!=97]
    df = df[df[x]!=98]
    df= df[df[x] != 99]
df['STATUS'] = df['STATUS'].apply(lambda x : x.replace('Cured', 'likely to cure').replace('Died','likely to die'))
df
```

```python
Q1 = df.AGE.quantile(0.25) 
Q2 = df.AGE.quantile(0.75)
Q1, Q2
```

```python
IQR = Q2 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q2 + 1.5*IQR
lower_limit, upper_limit
```

```python
df[(df.AGE <lower_limit) | (df.AGE > upper_limit)]
```

```python
df = df[(df.AGE >lower_limit) & (df.AGE < upper_limit)]
df.sample(20)
```

```python
x = df[['INTUBED', 'PNEUMONIA', 'AGE', 'DIABETES', 'ASTHMA', 'CARDIOVASCULAR',
       'OBESITY', 'TOBACCO', 'ICU']]
y = df.STATUS
```

```python
from sklearn.model_selection import train_test_split as ttt
X_train,X_test,y_train,y_test = ttt(x,y ,test_size = 0.1,  random_state = 10)
model = LogisticRegression(max_iter=5000,C = 20)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```
